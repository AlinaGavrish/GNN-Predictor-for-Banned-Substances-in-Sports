```python
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit import DataStructs
# Create an empty graph
# Convert the 'SMILES' column to string and replace any NaN values with an empty string
filtered_df3['SMILES'] = filtered_df3['SMILES'].astype(str).replace('nan', '')

# Create an empty graph
G = nx.Graph()

# Add nodes for drugs
for idx, row in filtered_df3.iterrows():
    smiles = row['SMILES']
    mol = Chem.MolFromSmiles(smiles)
    if mol:
        G.add_node(row['DrugBank ID'], name=row['Name'], groups=row['Drug Groups'], smiles=smiles, formula=row['Formula'], doping=row['Doping'])

# Calculate Morgan fingerprints for drugs
fingerprints = {}
for idx, row in filtered_df3.iterrows():
    mol = Chem.MolFromSmiles(row['SMILES'])
    if mol:
        fp = AllChem.GetMorganFingerprintAsBitVect(mol, 2)
        fingerprints[row['DrugBank ID']] = fp

# Calculate Tanimoto similarity between drugs
similarities = {}
for drug1_id, fp1 in fingerprints.items():
    for drug2_id, fp2 in fingerprints.items():
        if drug1_id != drug2_id:
            similarity = DataStructs.TanimotoSimilarity(fp1, fp2)
            if similarity > 0.7:  # Adjust threshold as needed
                if (drug2_id, drug1_id) not in similarities:  # Avoid duplicate edges
                    similarities[(drug1_id, drug2_id)] = similarity

# Add edges to the graph based on similarity
for (drug1_id, drug2_id), similarity in similarities.items():
    G.add_edge(drug1_id, drug2_id, similarity=similarity)

# Visualize the graph
pos = nx.spring_layout(G)  # Positions for all nodes
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='skyblue', font_size=10, font_weight='bold')
plt.show()
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the DrugBank page
url = 'https://go.drugbank.com/drugs/DB02327'

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content of the webpage
soup = BeautifulSoup(response.text, 'html.parser')

# Find the section containing Drug Categories information
drug_categories_section = soup.find('dt', id='drug-categories')

# Extract Drug Categories information
if drug_categories_section:
    # Find the list of drug categories
    category_list = drug_categories_section.find_next_sibling('dd').find('ul', class_='list-unstyled')

    # Extract category names
    if category_list:
        category_names = [category.text.strip() for category in category_list.find_all('a')]

        # Print the extracted category names
        print("Drug Categories:")
        for category_name in category_names:
            print(category_name)
    else:
        print("No drug categories found in the Drug Categories section.")
else:
    print("Drug Categories section not found on the webpage.")
```
